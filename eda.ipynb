# Python Implementation of Linear Regression

```python
# Import the libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("tips.csv")
```

```python
# To check the top 5 rows we run
df.head()
```

```python
# missing values
df.isnull().sum()
```

```python
# Separating independent and dependent variable
X=df.loc[:,["total_bill"]] # Independent variable
```

```python
type(X)
```

```python
y=df.loc[:,["tip"]] # dependent or target variable
```

```python
X.shape
```

```python
y.shape
```

```python
(80 * 244)/100 # Training, rest 49(20%) will go to testing
```

```python
# split the data into 80% and 20%
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test.count()
```

```python
# Model Training - X_train,y_train
# Model Evaluation - X_test and we get y_pred
# compare y_pred with y_test using cost function
# cost function value is lesser then our model is the best model
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
```

```python
# Let's train the data
# fit function will help you train the data
lr.fit(X_train,y_train)
```

```python
lr.intercept_ # intercept which is c.
```

```python
# slope of the line
lr.coef_ # m value
# y=mx+c
# tip = 0.09 * total_bill + 1.02
```

```python
y_test
```

```python
# Model testing
y_pred=lr.predict(X_test)
```

```python
# Now I have y_pred and y_test
# let's check the cost function
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_pred)
```

```python
# root mean squared error
root_mean_squared_error=np.sqrt(mse)
```

```python
print(f"The roor mean squared value is {root_mean_squared_error}")
```

```python
# 2 + 0.90
# 2-0.90
# Searalization - is a process of converting your model into a binary format
import joblib
joblib.dump(lr,"linear_reg.pkl")
```
